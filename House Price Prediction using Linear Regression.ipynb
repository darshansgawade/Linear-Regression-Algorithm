## Train and Test Model

```python
import numpy as np
import sklearn.model_selection

from sklearn.model_selection import train_test_split
```

```python
x = d[["area"]]
y = d[["price"]]
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0, train_size=0.75)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

```python
x = x_train.values
y = y_train.values
```

```python
from sklearn.linear_model import LinearRegression
```

```python
#creating object
Linear_R = LinearRegression()
```

```python
#training the model
Linear_R.fit(x_train,y_train)
```

```python
#Prediction
predicted_y=Linear_R.predict(x_test)
```

```python
predicted_y
```

```python
#slope
Linear_R.coef_
```

```python
#y parameter or C value
Linear_R.intercept_
```

```python
#predict values
Linear_R.predict([[4000]])
```

```python
#Area 15000 then price is
Linear_R.predict([[15000]])
```

```python
Linear_R.score(x_test,y_test)
```

```python
#mean_square_error
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test,predicted_y)
```

```python
#root mean square error
np.sqrt(mean_squared_error(y_test,predicted_y))
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(y_test,predicted_y)
```

```python
from sklearn.metrics import r2_score
```

```python
#r2 score
r2_score(y_test,predicted_y)
```
